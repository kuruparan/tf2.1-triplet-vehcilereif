```python
train_dataset, test_dataset = tfds.load(name="cars196", split=['train', 'test'], as_supervised=True)
```

```python
def _normalize_img(img, label):
    img = tf.cast(img, tf.float32) / 1.
    return (img, label)

#train_dataset, test_dataset = train_data, test_data
# Build your input pipelines
train_dataset = train_dataset.shuffle(1024).batch(32)
train_dataset = train_dataset.map(_normalize_img)

test_dataset = test_dataset.batch(32)
test_dataset = test_dataset.map(_normalize_img)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(filters=64, kernel_size=2, padding='same', activation='relu', input_shape=( 32, 32, 3)),
    tf.keras.layers.MaxPooling2D(pool_size=2),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Conv2D(filters=32, kernel_size=2, padding='same', activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=2),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(256, activation=None), # No activation on final dense layer
    tf.keras.layers.Lambda(lambda x: tf.math.l2_normalize(x, axis=1)) # L2 normalize embeddings
 #   tf.keras.layers.Dense(775, activation=None) # No activation on final dense layer

])
```

```python
# Compile the model
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.001),
   # loss=tf.keras.losses.CategoricalCrossentropy(),
     metrics=['accuracy'],
    loss=tfa.losses.TripletSemiHardLoss())
```

```python
# Train the network
history = model.fit(
    train_dataset,
    epochs=15)
```

```python
history = model.fit_generator(
    train_data,
    steps_per_epoch=34420 // 32,
    epochs=15,
    validation_data=test_data,
    validation_steps=8254 // 32
)
```

```python
results = model.predict(test_dataset)
```
